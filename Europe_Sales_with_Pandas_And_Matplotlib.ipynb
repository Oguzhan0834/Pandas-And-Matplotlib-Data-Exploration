```python
# Correcting some data types
EuropeSales["Unit Price"] = EuropeSales["Unit Price"].astype("int")
EuropeSales["Unit Cost"] = EuropeSales["Unit Cost"].astype("int")
EuropeSales["Total Revenue"] = EuropeSales["Total Revenue"].astype("int")
EuropeSales["Total Cost"] = EuropeSales["Total Cost"].astype("int")
EuropeSales["Total Profit"] = EuropeSales["Total Profit"].astype("int")
EuropeSales["Order Date"] = EuropeSales["Order Date"].astype("datetime64[ns]")
EuropeSales["Ship Date"] = EuropeSales["Ship Date"].astype("datetime64[ns]")
```

```python
# dropping duplicates
EuropeSales.drop_duplicates()
```

```python
# Pulling spesific rows by isin function
Two_Item_Type = ["Cereal" ,"Meal"]
EuropeSales["Item Type"].isin(Two_Item_Type)
EuropeSales[EuropeSales["Item Type"].isin(Two_Item_Type)].sort_index(ascending=True)
```

```python
# Pulling speesific row by str.contains 
EuropeSales[EuropeSales["Sales Channel"].str.contains("Online")]
```

```python
# Setting index
EuropeSales2 = EuropeSales.set_index("Country")
EuropeSales2
```

```python
# Countries that made more than 1000000 money 
EuropeSales[EuropeSales["Total Revenue"] > 1000000].sort_values(by=["Item Type"])
```

```python
# Beverages sales more than 6000 and as of 01/01/2013
is_beverages = EuropeSales["Item Type"] == "Beverages"
is_sales_over_6000 = EuropeSales["Units Sold"] > 6000
is_ship_date_as_of_2012 = EuropeSales["Ship Date"] > "12/31/2012"
```

```python
EuropeSales[is_beverages & is_sales_over_6000 & is_ship_date_as_of_2012]
```

```python
# Sales between 2015 and 2017
EuropeSales[EuropeSales["Ship Date"].between("01/01/2015" ,"12/31/2017")]
```

```python
# Shipping time
EuropeSales["Shipping Time"] = EuropeSales["Ship Date"] - EuropeSales["Order Date"]
```

```python
# Grouping by categories
categories = EuropeSales.groupby("Item Type")
```

```python
# Looking at total values by categories
categories.sum()
```

```python
# Looking at average values by categories
categories.mean()
```

```python
# Looking at sales values by aggregation  
EuropeSales.groupby("Item Type").agg({"Total Profit":["sum","mean","min","max","std"]})
```
